```python
import pandas as pd
import numpy as np
from word2number import w2n  # Necesitarás instalar esta librería: pip install word2number

# Leer el archivo CSV con el delimitador correcto
df = pd.read_csv('HR RAW DATA CELDAS.csv', delimiter=';')

# Función para convertir palabras a números
def convert_to_number(age):
    if pd.isna(age) or age == '':
        return np.nan
    try:
        # Intentar convertir directamente a número
        return int(float(age))
    except ValueError:
        try:
            # Convertir palabra a número
            return w2n.word_to_num(age.lower())
        except ValueError:
            return np.nan

# Aplicar la función a la columna 'Age'
df['Age'] = df['Age'].apply(convert_to_number)

# Verificar el resultado
print(df.sample(50))

# Sobrescribir el archivo CSV original
df.to_csv('HR RAW DATA CELDAS.csv', index=False, sep=';')

```

```python
import pandas as pd
import numpy as np

def convert_daily_rate(file_path):
    try:
        # Leer el archivo CSV con el delimitador correcto
        df = pd.read_csv(file_path, delimiter=';')
    except Exception as e:
        print(f"Error al leer el archivo {file_path}: {e}")
        return
    
    try:
        # Verificar si la columna 'DailyRate' existe
        if 'DailyRate' in df.columns:
            # Reemplazar valores vacíos con NaN
            df['DailyRate'] = df['DailyRate'].apply(lambda x: np.nan if x == '' else x)
            # Convertir valores a numéricos, manejar no numéricos como NaN
            df['DailyRate'] = pd.to_numeric(df['DailyRate'], errors='coerce')
            # Convertir la columna a tipo entero
            df['DailyRate'] = df['DailyRate'].astype('Int64')
        else:
            print(f"Advertencia: La columna 'DailyRate' no se encontró en el archivo {file_path}.")
            return
    except Exception as e:
        print(f"Error al procesar la columna 'DailyRate': {e}")
        return
    
    try:
        # Guardar el DataFrame en el mismo archivo CSV para sobrescribirlo
        df.to_csv(file_path, index=False, sep=';')
    except Exception as e:
        print(f"Error al escribir el archivo {file_path}: {e}")
        return

    print("Archivo procesado y guardado exitosamente.")
    print(df.sample(5))

# Ruta del archivo CSV
file_path = 'HR RAW DATA CELDAS.csv'

# Procesar el archivo CSV
convert_daily_rate(file_path)
```

```python
import pandas as pd
import re

# Supongamos que el archivo CSV está cargado en la variable 'file_path'
file_path = 'HR RAW DATA CELDAS.csv'

# Cargar el archivo CSV
df = pd.read_csv(file_path)

# Función para estandarizar los títulos de las columnas al formato "CamelCase"
def standardize_column_name(col_name):
    # Insertar un espacio antes de cada mayúscula que no esté precedida por otro mayúscula (para casos tipo PascalCase y camelCase)
    col_name = re.sub(r'([a-z])([A-Z])', r'\1 \2', col_name)
    col_name = re.sub(r'([A-Z])([A-Z][a-z])', r'\1 \2', col_name)
    # Capitalizar cada palabra y unirlas sin espacios
    words = col_name.split()
    return ''.join(word.title() for word in words)



# Aplicar la estandarización a los nombres de las columnas
df.columns = [standardize_column_name(col) for col in df.columns]

# Guardar el DataFrame modificado en un nuevo archivo CSV
output_file_path = 'HR RAW DATA CELDAS.csv'
df.to_csv(output_file_path, index=False)

print(f"Se han estandarizado los nombres de las columnas y guardado en '{output_file_path}'.")
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import words

# Descargar los datos necesarios de NLTK si no están disponibles
nltk.download('punkt', quiet=True)
nltk.download('words', quiet=True)

# Supongamos que el archivo CSV está cargado en la variable 'file_path'
file_path = 'HR RAW DATA CELDAS.csv'

# Cargar el archivo CSV
df = pd.read_csv(file_path)

# Función para estandarizar los títulos de las columnas al formato "CamelCase"
def standardize_column_name(col_name):
    words = []
    current_word = col_name[0].title()  # Capitalizar el primer carácter
    
    for char in col_name[1:]:
        if char.isupper() and (current_word[-1].islower() or (len(current_word) > 1 and current_word[-2].islower())):
            # Detected a transition from lower to upper case, start a new word
            words.append(current_word)
            current_word = char
        else:
            current_word += char
    
    words.append(current_word)  # Add the last accumulated word
    return ''.join(word.capitalize() for word in words)

# Aplicar la estandarización a los nombres de las columnas
df.columns = [standardize_column_name(col) for col in df.columns]

# Guardar el DataFrame modificado en un nuevo archivo CSV
output_file_path = 'HR RAW DATA CELDAS.csv'
df.to_csv(output_file_path, index=False)

print(f"Se han estandarizado los nombres de las columnas y guardado en '{output_file_path}'.")
```
